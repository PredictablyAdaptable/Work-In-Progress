```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# x_train is the input variable, 
# y_train is the target
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(x_train.shape)
m = x_train.shape[0]
print(m)
```

```python
i = 1

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker = 'x', c= 'r')
plt.title('Housing Prices')
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 100
b = 100

# Creating UDF for model(linear reg) parameters are x = independent variable, w = model weight/slope, b = bias/intercept
def compute_model_output(x,w,b):
    m = x.shape[0]
    f_wb = np.zeros(m) #creates a numpy array filled with zeros, where size of the array m, essentially initating the np.array in prep of for loop
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb

# this function essentially returns a np.array of predicted values for all range(m)
```

```python
# useing the funciton (tmp_f_wb contains an np.array of predicted values for independent values in np.array x_train)
tmp_f_wb = compute_model_output(x_train, w, b)

#plot the data points
plt.plot(x_train, tmp_f_wb, c = 'b', label = 'Our Prediction')

#Plot the data points
plt.scatter(x_train, y_train, marker = 'x', c = 'r', label= 'Actual Values')

#Visualizaiton Aesthetics
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
