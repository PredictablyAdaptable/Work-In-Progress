#### Binary Search
 


1. The Searching Process
    1. Simple search
        1. Start from first element work your way one by one until you find the value
    2. Binary Search
        1. Input is a sorted list. If the element you're looking for is in the list, binary search returns the position where it's located, otherwise null.
        2. You guess the middle index and eliminate half the remaining numbers every time

            1. Since log is the flip of exponents i.e. 10^2 = 100, then log_10 of 100 = 2
            2. Binary search algo is
                1. Log_2 len(list) = number of steps to find index value

### Chapter 2: Selection Sort

  
1. Arrays -> its like standing in line, if you want to insert a person on the nth place of the array, then everyone left of the line needs to move.
2. Linked lists -> items are not stored in a line, but in a bunch of different rooms ('nodes')
    1. The nodes 'rooms' contain
        1. Data -> the value or information the node holds
        2. Pointer -> the address of the next node in the sequence
    2. Thus linked lists still has an order, only difference being when inserting a new data point, you don't need to shift every data point to the left of the insertion point. You just have to adjust the pointer of the node just before the insertion point.
    3. Advantages of linked lists over Arrays
        1. Sequential Access -> Memory doesn't have to be stored together
        2. Don't need to shift data for inserting data in to a linked list
    4. Disadvantages of linked list over Arrays
        1. To go to the nth element of a linked list, you still have to start from the beginning, and go to different rooms 1 by 1 until you reach the room you want.
            1. Random Access -> In arrays, you know exactly what the nth item is without going through all the items before the nth item

1. Nested Functions and Stack
    1. Call Stack -> a stack data structure that keeps track of active function calls in a program. It manages the order of function execution and ensures that each function's execution context is preserved
    2. Stack Frame -> a block of memory allocated on the call stack for each function call.
    3. Function Call -> the process of invoking a function which involves creating a new stack frame
    4. Process of Stack when Processing a Function 
        1. Function Invocation (Call)
            1. Stack Frame Created
                1. A new memory is allocated to the stack frame
                2. Written in the memory
                    1. Return address -> where in the code the function was called, so the program can return to that point after function is executed
                    2. Function params -> arguments passed in function
                    3. Local variables -> variables declared with the function
                    4. Saved registers
                        1. Registers are small, fast storage locations within the CPU used to hold data temporarily during execution
                        2. They store operands for arithmetic operations, address memory access and other critical control information
                        3. Thus you save the registers state before the function was called so you can return to the original state after the function call.
            2. Pushing onto the Call Stack -> involves adding a new stack frame to the top of the call stack to represent the call function's execution context
            3. Transfer of Control -> process by which the CPU begins executing the called function's code instead of continuing with the current function
        2. Function Execution
            1. Accessing local data -> the function accesses its local variables and parameters from its stack frame
            2. Nested function calls -> same process repeats if there is a nested function
        3. Function Return
            1. When a function complete's its execution
                1. Pop stack frame
                2. Return address
                3. Transfer of control
                4. Cleanup -> any resources allocated within the function are cleaned up, and previous execution context is restored
 


1. Simple analogy
    1. Function context
        1. You have a function a
        2. Within function a there is function b and function c, in that order
        3. Within function b there is function b.1
    2. This is how function a is executed with stack
        1. Function a called on to memory
        2. Mid way executing function a, function b is encountered
        3. Function b is pushed on top of function a
        4. Mid way executing function b, function b.1 is encountered
        5. Function b.1 is pushed on top of function b
        6. Now the stack consists of from top to bottom, function b.1, b, a
        7. Function b.1 is executed, then popped off
        8. Now stack consists of from top to bottom b, a
        9. Function b is executed, then popped off
        10. Function a continues to run
        11. Function c is encountered
        12. Function c is pushed on top of function a
        13. Function c is executed and popped off
        14. Function a is executed and finished
2. Same thing happens in recursions
